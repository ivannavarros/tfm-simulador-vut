```python
# ============================================================
# CELDA 3 — Carga de datos
# ============================================================
import os

BASE = os.path.abspath(os.path.join(os.getcwd(), '..'))

PATH_MADRID = os.path.join(BASE, 'data', 'raw', 'madrid', 'listings.csv.gz')
PATH_BCN    = os.path.join(BASE, 'data', 'raw', 'barcelona', 'listings.csv.gz')

# Carga
madrid    = pd.read_csv(PATH_MADRID, compression='gzip', low_memory=False)
barcelona = pd.read_csv(PATH_BCN, compression='gzip', low_memory=False)

# Añadimos columna ciudad
madrid['ciudad']    = 'madrid'
barcelona['ciudad'] = 'barcelona'

print(f"Madrid:    {madrid.shape[0]:,} anuncios  |  {madrid.shape[1]} columnas")
print(f"Barcelona: {barcelona.shape[0]:,} anuncios  |  {barcelona.shape[1]} columnas")
```

```python
# ============================================================
# CELDA 4 — Exploración inicial
# ============================================================

# Unimos ambos datasets
df = pd.concat([madrid, barcelona], ignore_index=True)

print(f"Dataset combinado: {df.shape[0]:,} anuncios | {df.shape[1]} columnas")
print(f"\n--- Distribución por ciudad ---")
print(df['ciudad'].value_counts())

print(f"\n--- Columnas disponibles ---")
for i, col in enumerate(df.columns):
    print(f"  {i+1:02d}. {col}")
```

```python
# ============================================================
# CELDA 5 — Selección de variables relevantes
# ============================================================

COLUMNAS = [
    # Identificación
    'id', 'ciudad',
    
    # Localización
    'neighbourhood_cleansed', 'latitude', 'longitude',
    
    # Características del inmueble
    'property_type', 'room_type', 'accommodates',
    'bedrooms', 'beds', 'bathrooms_text',
    
    # Precio
    'price',
    
    # Disponibilidad
    'availability_30', 'availability_60',
    'availability_90', 'availability_365',
    'minimum_nights',
    
    # Reputación
    'number_of_reviews', 'number_of_reviews_ltm',
    'review_scores_rating', 'review_scores_cleanliness',
    'review_scores_location', 'review_scores_value',
    'reviews_per_month',
    
    # Host
    'host_is_superhost', 'host_response_rate',
    'host_acceptance_rate', 'host_listings_count',
    'hosts_time_as_host_years',
    
    # Variables objetivo
    'estimated_occupancy_l365d',
    'estimated_revenue_l365d',
    
    # Reserva instantánea
    'instant_bookable',
]

df_sel = df[COLUMNAS].copy()

print(f"Dataset reducido: {df_sel.shape[0]:,} filas | {df_sel.shape[1]} columnas")
print(f"\n--- Valores nulos por columna ---")
nulos = df_sel.isnull().sum()
pct   = (nulos / len(df_sel) * 100).round(1)
resumen_nulos = pd.DataFrame({'nulos': nulos, '%': pct})
print(resumen_nulos[resumen_nulos['nulos'] > 0].sort_values('%', ascending=False))
```

```python
# ============================================================
# CELDA 6 — Limpieza inicial
# ============================================================

print(f"Anuncios antes de limpieza: {len(df_sel):,}")

# 1. Eliminamos filas sin precio (no podemos trabajar sin él)
df_clean = df_sel.dropna(subset=['price']).copy()
print(f"Tras eliminar sin precio:   {len(df_clean):,}")

# 2. Limpiamos columna price (viene como '$1,234.00')
df_clean['price'] = (
    df_clean['price']
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

# 3. Eliminamos prices imposibles (0 o >5000)
df_clean = df_clean[(df_clean['price'] > 0) & (df_clean['price'] <= 5000)]
print(f"Tras filtrar precios:       {len(df_clean):,}")

# 4. Eliminamos columna con demasiados nulos
df_clean = df_clean.drop(columns=['hosts_time_as_host_years'])

# 5. Limpiamos bathrooms_text → número
df_clean['bathrooms'] = (
    df_clean['bathrooms_text']
    .str.extract(r'(\d+\.?\d*)')
    .astype(float)
)
df_clean = df_clean.drop(columns=['bathrooms_text'])

print(f"\n--- Distribución por ciudad tras limpieza ---")
print(df_clean['ciudad'].value_counts())

print(f"\n--- Estadísticas de precio por ciudad ---")
print(df_clean.groupby('ciudad')['price'].describe().round(2))
```

```python
# ============================================================
# CELDA 9 — Carga y diagnóstico de los cuatro datasets
# ============================================================

import os
import pandas as pd

BASE = os.path.abspath(os.path.join(os.getcwd(), '..'))
RAW  = os.path.join(BASE, 'data', 'raw')

datasets = {
    'madrid_sep':    os.path.join(RAW, 'madrid',    'sep2025', 'listings.csv.gz'),
    'madrid_jun':    os.path.join(RAW, 'madrid',    'jun2025', 'listings.csv.gz'),
    'barcelona_sep': os.path.join(RAW, 'barcelona', 'sep2025', 'listings.csv.gz'),
    'barcelona_jun': os.path.join(RAW, 'barcelona', 'jun2025', 'listings.csv.gz'),
}

for nombre, ruta in datasets.items():
    df_tmp = pd.read_csv(ruta, compression='gzip', low_memory=False)
    tiene_precio = 'price' in df_tmp.columns
    nulos_precio = df_tmp['price'].isnull().sum() if tiene_precio else 'N/A'
    pct_nulos    = f"{nulos_precio/len(df_tmp)*100:.1f}%" if tiene_precio else 'N/A'
    scraped      = df_tmp['last_scraped'].unique()[:2] if 'last_scraped' in df_tmp.columns else 'N/A'
    
    print(f"\n📊 {nombre}")
    print(f"   Anuncios:      {len(df_tmp):,}")
    print(f"   Columnas:      {df_tmp.shape[1]}")
    print(f"   Tiene precio:  {tiene_precio}")
    print(f"   Nulos precio:  {nulos_precio} ({pct_nulos})")
    print(f"   Fecha scrape:  {scraped}")
```

```python
# ============================================================
# CELDA 10 — Carga y combinación final
# ============================================================

def cargar(ruta, ciudad, fecha):
    df = pd.read_csv(ruta, compression='gzip', low_memory=False)
    df['ciudad'] = ciudad
    df['fecha_scrape'] = fecha
    return df

madrid_sep    = cargar(os.path.join(RAW, 'madrid',    'sep2025', 'listings.csv.gz'), 'madrid',    'sep2025')
madrid_jun    = cargar(os.path.join(RAW, 'madrid',    'jun2025', 'listings.csv.gz'), 'madrid',    'jun2025')
barcelona_sep = cargar(os.path.join(RAW, 'barcelona', 'sep2025', 'listings.csv.gz'), 'barcelona', 'sep2025')
barcelona_jun = cargar(os.path.join(RAW, 'barcelona', 'jun2025', 'listings.csv.gz'), 'barcelona', 'jun2025')

# Combinamos
df = pd.concat([madrid_sep, madrid_jun, barcelona_sep, barcelona_jun], ignore_index=True)

print(f"Dataset combinado: {df.shape[0]:,} anuncios | {df.shape[1]} columnas")
print(f"\n--- Anuncios por ciudad y fecha ---")
print(df.groupby(['ciudad', 'fecha_scrape']).size().reset_index(name='anuncios'))

# Verificamos columnas clave
cols_clave = ['price', 'estimated_occupancy_l365d', 'estimated_revenue_l365d',
              'neighbourhood_cleansed', 'room_type', 'accommodates', 'bedrooms']
print(f"\n--- Columnas clave presentes ---")
for col in cols_clave:
    presente = col in df.columns
    print(f"   {'✅' if presente else '❌'} {col}")
```
